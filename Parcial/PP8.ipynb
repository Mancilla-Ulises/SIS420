## Datos de Cáncer de Mama de Wisconsin (Breast Cancer Wisconsin - Original)

El conjunto de datos Breast Cancer Wisconsin (Original) contiene características calculadas a partir de imágenes de aspirados con aguja fina (FNA) de masas mamarias. El objetivo es predecir si una masa es benigna o maligna.

**Dataset:** [Breast Cancer Wisconsin (Original) - UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/breast+cancer+wisconsin+(original))

## 2. Carga y Selección de Datos

Se carga el archivo `.data` sin encabezado y se asignan los nombres de columnas manualmente. Los valores faltantes en este dataset están representados como `'?'`.

| Columna (índice) | Nombre             | Descripción |
|------------------|--------------------|-------------|
| 0                | `sample_code`      | Número de código de muestra |
| 1                | `clump_thickness`  | Grosor del agregado celular (1-10) |
| 2                | `cell_size_uniformity` | Uniformidad del tamaño celular (1-10) |
| 3                | `cell_shape_uniformity` | Uniformidad de la forma celular (1-10) |
| 4                | `marginal_adhesion` | Adhesión marginal (1-10) |
| 5                | `single_cell_size` | Tamaño de célula epitelial única (1-10) |
| 6                | `bare_nuclei`      | Núcleos desnudos (1-10) |
| 7                | `bland_chromatin`  | Cromatina blanda (1-10) |
| 8                | `normal_nucleoli`  | Nucléolos normales (1-10) |
| 9                | `mitoses`          | Mitosis (1-10) |
| 10               | `class`            | **Variable objetivo:** 2 = benigno, 4 = maligno |

**Variable objetivo (col. 10):** `class` (2 = benigno, 4 = maligno)

```python
import pandas as pd

# Nombres de columnas (el archivo no tiene encabezado)
column_names = [
    'sample_code', 'clump_thickness', 'cell_size_uniformity', 
    'cell_shape_uniformity', 'marginal_adhesion', 'single_cell_size',
    'bare_nuclei', 'bland_chromatin', 'normal_nucleoli', 'mitoses', 'class'
]

dt = pd.read_csv('../Database/PP8_breast-cancer-wisconsin_clean.data',
    header=None,
    names=column_names,
    skipinitialspace=True # elimina espacios al inicio de cada campo
)

# Limpiar espacios en columnas de texto
dt.columns = dt.columns.str.strip()

# Eliminar columna sample_code (identificador, no es predictora)
columnas_a_excluir = ['sample_code']
dt = dt.drop(columns=columnas_a_excluir)

print(dt.columns.tolist())
print(f"Total filas: {len(dt)}")
print(f"Valores NaN: {dt.isnull().sum().sum()}")
print(f"Valores 0: {(dt == 0).sum().sum()}")
```

```python
numeric_cols = dt.select_dtypes(include=['number']).columns.tolist()
categorical_cols = dt.select_dtypes(include=['object']).columns.tolist()
```

```python
# Verificar que no hay valores faltantes ni ceros
print("=== Verificación de datos limpios ===")
print(f"Valores NaN por columna:\n{dt.isnull().sum()}")
print(f"\nValores 0 por columna:\n{(dt == 0).sum()}")

# Todas las columnas son numéricas en este dataset
numeric_cols = dt.select_dtypes(include=['number']).columns.tolist()
categorical_cols = dt.select_dtypes(include=['object']).columns.tolist()

print(f"\nColumnas numéricas: {numeric_cols}")
print(f"Columnas categóricas: {categorical_cols}")

# Como el dataset ya está limpio, solo escalamos para KNN
scaler = StandardScaler()
X_num = scaler.fit_transform(dt[numeric_cols])

# KNN sobre las columnas numéricas (k=5 vecinos)
knn = NearestNeighbors(n_neighbors=6, metric='euclidean')  # 6 porque incluye la propia fila
knn.fit(X_num)
distancias, indices = knn.kneighbors(X_num)  # precalcular vecinos para todas las filas

print(f"\n✓ Datos listos para modelado.")
print(f"¿Quedan NaN en todo dt? {dt.isnull().sum().sum()}")
print(f"¿Quedan 0 en todo dt? {(dt == 0).sum().sum()}")
print("\nTipos de datos finales:")
print(dt.dtypes.value_counts())
print("\nPrimeras 5 filas:")
print(dt.head())
```
